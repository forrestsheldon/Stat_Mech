####Microscopic Model

A simple kinetic model of noninteracting particles yields,
$$PV = N m\overline{v_x^2}$$
which suggests the average translational kinetic energy in each direction
is $\overline{\frac{1}{2}mv_x^2} = \frac{1}{2}kT$ and thus the total translational kinetic
energy is $$\overline{K} =U = \frac{3}{2}kT.$$
To get an idea of the energy scales we are talking about here, at room temperature
$$kT = (1.38\times 10^{-23}J/K)(300K) = 4.14\times 10^{-21}J\approx \frac{1}{40}eV$$
We can also get the approximate speed of a molecule in the gas,
$$v_{rms} \equiv \sqrt{\overline{v^2}} = \sqrt{\frac{3kT}{m}}$$
(Note that this is slightly larger than $\bar{v}$.) For a nitrogen molecule at room temperature,
this yields,

```python
np.sqrt(3*k_B*293/(0.028/N_A))
```

It's no accident that this is near to the speed of sound, but that's a non-equilibrium class.  It is also worth noting (and super cool) that while the
derivation relies on the particles not interacting, if the gas becomes dense the ideal gas law
breaks down in such a way to conserve the expression for the total energy and so $U = \frac{3}{2}kT$ is true
for dense gases, liquids and sometimes solids!

is true at high temperatures.  At room temperatures in a gas, vibrational degrees of freedom (which

### Heat and Work

The first rule of thermodynamics is an expression of energy conservation and gives heat a definite form in relation to energy, which gets us on our way to defining temperature in terms of more exact arguements. We classify the manner in which the thermal energy can change into two categories, heat and work. So
$$\Delta U = Q + W$$

Heat is any spontaneous flow of energy from one object to another caused by a difference in temperature.

Work is any other transfer of energy.  Because these refer only to exchanges of energy, there is no heat or work function that describes the amount of heat or work in a system.  We can refer only to the total energy in a system and the amount of heating or work done on it.

Heat was originally measured in calories which is the amount of heat required to raise the temperature of a gram of water by 1$^\circ$ Celcius. Joule showed that 1 calorie is approximately 4.2J and today it is defined at 4.186J.

### Compression Work

When we compress a system, we can attempt to use our classical definition of work in terms of force.  However, in order to use the pressure to characterize the force, we must assume that the change is $quasistatic$, meaning that the change in the volume is 
sufficiently slow that the gas has time to continually equilibrate, which means that the pressure is well-defined throughout the gas.  As at room temperature the gas molecules are moving faster than the speed of sound, we would need to compress at near that speed to compress faster than the gas could respond.

If we have all of this, we get $W = -P\Delta V$.

#### Ideal Gas

In the ideal gas, we can consider two types of compression work:  Slow compression, where the gas exchanges heat with its surroundings and keeps the temperature constant ie $isothermal$ compression, and fast compression, which is so fast that no heat is exchanged with the surroundings ie $adiabatic$ compression.

##### Isothermal Compression

The inverse volume dependence yields, as we would expect,
$$W = N k_B T \ln{\frac{V_i}{V_f}}$$
So each factor that we compress a gas by increases the energy by a constant amount proportional to $N k_B T$.  So the same work is required to half the volume of a gas, no matter what the starting volume.  As the thermal energy doesn't change, we have $Q = -W$. To compress a mole of ideal gas to half it's volume, we find

```python
N_A*k_B*T_room*np.log(2)
```

```python
fig2 = figure()
ax2 = fig2.add_axes([0.1, 0.1, 0.8, 0.8])

V = linspace(0.001, 0.05)
ax2.plot(V, N_A*k_B*293*np.log(0.001/V),'r')
ax2.plot(V, N_A*k_B*100*np.log(0.001/V),'g')
ax2.plot(V, N_A*k_B*400*np.log(0.001/V),'b')
ax2.set_xlabel('Volume $m^3$')
ax2.set_ylabel('Work $(J)$')
ax2.set_title('Work Done on a Liter of Gas During its Expansion')
```

#####Adiabatic Compression

In adiabatic compression, $Q=0$ so $\frac{f}{2}N k dT = -P dV$ and utilizing the ideal gas law,
$$\frac{f}{2}\frac{dT}{T} = -\frac{dV}{V}$$
Integrating this and exponentiating we obtain
$$V_f T^{f/2}_f = V_i T^{f/2}_i$$
or
$$V^\gamma P = constant$$
where $\gamma$ is the adiabatic exponent, $\gamma = (f+2)/f$

```python
gamma_mono=(3.+2)/3
```

```python
t = linspace(0, 400, 100)
v = linspace(0.005, 0.1, 100)
T,V = meshgrid(t,v)
P = pressure_ideal(N_A, T, V)

V_proc = linspace(0.01,0.09,50)
P_adiabat = N_A*k_B*350*0.01**(2./3)/V_proc**(gamma_mono)
T_adiabat = 350*(0.01/V_proc)**(2./3)

T_iso = [350 for i in range(50)]
P_iso = N_A*k_B*350./V_proc
T_contour = zeros(50)

fig2 = figure(figsize=(12,8))
ax2 = fig2.add_subplot(1, 1, 1, projection = "3d")

ax2.plot_surface(T, V, P, rstride=7, cstride=7, alpha=0.05)
ax2.plot(T_adiabat,V_proc,P_adiabat,'r',linewidth=2)
ax2.plot(T_iso,V_proc,P_iso,'b',linewidth=2)
ax2.plot(T_contour, V_proc, P_adiabat, 'r')
ax2.plot(T_contour, V_proc, P_iso, 'b')


ax2.set_zlim3d(0,4e5)
ax2.set_title("The Ideal Gas Law ($1mol$)")
ax2.set_xlabel('Temperature $(K)$')
ax2.set_ylabel('Volume $(m^3)$')
ax2.set_zlabel('Pressure $(N/m^2)$')
ax2.view_init(30, 30)
```

Here we see why an adiabat is steeper than an isotherm when looked at on a PV diagram.  Physically, this is because when we allow a gas to expand isothermally, it will be heated by it's surroundings and thus its pressure will be larger than a gas expanding adiabatically.

###Heat Capacities

The heat capacity is the amount of heat required to raise the temperature of a system per degree temperature change
$$C = \frac{Q}{\Delta T} = \frac{\Delta U - W}{\Delta T}$$
This quantity is extensive and so we define an intensive quantity, the specific heat capacity
$$c = \frac{C}{m}.$$
This definition is ambiguous as we have not specified how the head is added.  We define two heat capacities:
The heat capacity at constant volume, such that $W=0$ and 
$$C_v = \left(\frac{\partial U}{\partial T}\right)_V$$
We can think of this as the energy capacity of a system as it is the energy required to raise the temperature of the system. For a gram of water, the energy capacity is $4.2J/^\circ C$.

Uncontained systems will expand as they are heated and thus excess heat will be required to supply the work done by the expansion. For compression work,
$$C_P = \left(\frac{\partial U}{\partial T}\right)_P + P\left(\frac{\partial V}{\partial T}\right)_P$$
For most solids and liquids we may neglect the second term (example?) but for gasses it is necessary.

If an object holds thermal energy only in quadratic degrees of freedom, then
$$C_V = \frac{fNk_B}{2}$$
For a monatomic ideal gas, the heat capacity per mol is thus $\frac{3}{2}R = 12.5 J/K.$ For a solid at high temperatures, it is $3R = 25J/K$ which is the Dulong Petit rule.  At low temperatures, all of the degrees of freedom freeze out so that $C_v \to 0$.

For an ideal gas,
$$C_P = C_V + Nk = C_V + nR$$

####Latent Heat

During a phase tranformation, we can add heat without changing the temperature and thus the heat capacity is infinite $C\to\infty$. We use the latent heat to quantify the heat necessary to completely melt or boil the substance.
$$L = \frac{Q}{m}$$
For water, the latent heat of melting is 333 J/g or 80 cal/g and the latent heat of boiling is 2260 J/g or 540 cal/g. (Recall that it takes 100 cal/g to raise the temperature of water from $0 ^\circ C$ to $100 ^\circ C$.)
####Enthalpy

For constant pressure processes, keeping track of the compression work can be a pain, so we create a new quantity call the enthalpy
$$H = U + PV$$
This is the total energy required to create a system in a constant pressure environment. By examing differentials, we find that for constant pressure processes,
$$\Delta H = Q + W_{other}$$
where $W_{other}$ is work other than compression-expansion work so if no work is done, the change in enthalpy is the heat added or removed from the system (hence the H).

We can recognize
$$C_P = \left(\frac{\partial H }{\partial T}\right)_P$$
so $C_P$ can be considered the "enthalpy capacity."

#####Enthalpy Examples
1. Chemistry books quote $\Delta H$ for a variety of processes.  For boiling a mol of water, $\Delta H = 40,660J$.  So for one gram, $\Delta H = (40,660 J)/18 g = 2260J$ which is the latent heat.  As it requires $N_A k_B T = RT = 3100J$ of work to expand in the atmosphere, only 37,560J go into the thermal energy of the gas.

2. For $H_2 + \frac{1}{2} O_2 \to H_2 O$ the change in enthalpy per mol is -286kJ. This is referred to as the "enthalpy of formation". If we burn a mol of hydrogen, this is the energy we get out.

####Heat Conduction
When two systems are in contact through a medium heat will be conducted according to the Fourier heat conduction law,
$$\frac{Q}{\Delta t} = -k_t A \frac{dT}{dx}.$$
$k_t$ is the thermal conductivity and varies between materials by 4 orders of magnitude.  Some examples:

* 0.026 air
* 0.08 wood
* 0.6 water
* 0.8 glass
* 80 iron
* 400 copper

####Conductivity of an Ideal Gas

To get an approximation of the mean free path, we consider that a collision will occur when a particle comes within $2r$ of another.  When the volume swept out is equal to the average volume per molecule,
$$\pi(2r)^2l \approx \frac{V}{N}$$
$$l\approx \frac{1}{4\pi r^2}\frac{V}{N}$$
Given an approximate radius of 1.5 angstroms and treating air as an ideal gas at room temperature such that $\frac{V}{N} = \frac{kT}{P} = 4\times 10^{-26} m^3$, we get an approximate mean free path of

```python
l_air = 1/(4*np.pi*(1.5e-10)**2)*4e-26
print "The mean free path in air is approximately %.2g meters" % l_air
```

This yields a time between collisions of approximately

```python
1.5e-7/500
```

To derive Fourier's heat conduction law, consider a small region of gas where the temperature increases in the x-direction and calculate the heat that flows across a barrier in the middle. Consider a region of length $l$ to either side of the barrier.  If $\Delta t$ is the time between collisions, then in this time half of the particles in the left side will pass to the right and vice versa.  If $U_L$ is the energy in the region on the left and $U_R$ is the energy on the right, then the energy that flows across the barrier to the right is
$$Q = \frac{1}{2}(U_L - U_R) = -\frac{1}{2}(U_R - U_L) = -\frac{1}{2}C_V(T_R - T_L) = -\frac{1}{2}C_Vl \frac{dT}{dx}$$
This gives us an estimate of the thermal conductivity $k_t = \frac{1}{2}\frac{C_V l}{A \Delta t} = \frac{1}{2}\frac{C_V l^2}{A l \Delta t} = \frac{1}{2}\frac{C_v}{V}l\bar{v}$ The heat capacity of the gas per unit volume is $\frac{C_V}{V} = \frac{f}{2}\frac{Nk}{V} = \frac{f}{2}\frac{P}{T}.$ This temperature dependence cancels with the linear dependence in $l$ such that the thermal conductivity depends on the temperature only through $\bar{v}\propto \sqrt{T}$ (and possibly $f$ but only weakly in a narrow temperature range).

####Viscosity

Momentum can also spread through a fluid at the molecular level. Consider to parallel plates separated by a fluid.  If we move one of the plates, at low velocities the fluid will experience laminar flow.  Resistance to this flow is called viscosity. We can guess that the force is proportional to the common area of the plates and difference in their velocities and inversely proportional to the separation.  The constant of proportionality is called the coefficient of viscocity.
$$\frac{|F_x|}{A} = \eta \frac{du_x}{dz}$$
The force per unit area is shear stress and acts parallel to a surface rather than perpendicular like a pressure.  $\eta$ varies over many orders of magnitude and with temperature. Some examples:

* 0.0018 $Pa\cdot s$ water at $0^\circ C$
* 0.00028 $Pa\cdot s$ water at $100^\circ C$
* 0.25 $Pa\cdot s$ low viscocity motor oil (SAE 10) at room temperature
* 19 $\mu Pa\cdot s$ air at room temp

The viscocity of an ideal gas is independent of the pressure and increases with temperature. 

####Diffusion
In a fluid, particles will flow to oppose any difference in concentration.
$$J_x = -D\frac{dn}{dx}$$
This is Fick's Law and $D$ is the diffusion coefficient which depends on what molecule is diffusing and what it is diffusing through. 

##The Second Law

###Multiplicity, Macrostate and Microstate

By specifying all of the degrees of freedom associated with a system (or equivalently a point in phase space) we have chosen a **microstate** of the system.

Alternatively, if we specify only a constraint on the degrees of freedom of the system such as the total energy or momentum, (or equivalently a region of phase space) we have chosen a **macrostate** of the system.

The number of microstates that satisfy the constraint specifying a macrostate is called the **multiplicity** of the macrostate.

###Einstein Solid

A simple model of a solid that we can use to investigate the properties of these systems consists of N harmonic oscillators (one for each atom of $N/3$ in each direction) all of the same frequency.  The microstate is specified by giving the number of units of energy in each oscillator. The multiplicity of a state with $q$ units of energy is thus:
$$\Omega (N,q) = {q+N-1\choose N-1}= \frac{(q+N-1)!}{q!(N-1)!}$$
###Interacting Systems
Consider two Einstein solids that are **weakly coupled** such that exchanging units of energy between the solids is much slower than the exchange of energy between the oscillators within each. We specify a macrostate by the number of units of energy that lie in each solid. The multiplicity of the combined system is then just the product of the multiplicities of each,
$$\Omega_{AB} = \Omega_A\Omega_B.$$
We now make the **fundamental assumption of statistical mechanics**, that in an isolated system in thermal equilibrium, all accessible microstates (ie all that are compatible with the macrostate) are equally probable. The system then moves randomly between the many accessible microstates (at least on the timescales over which we can make measurements). Much more to say here.

The probability of a macrostate is then $\frac{\Omega_{macrostate}}{\Omega_{all}}$. We find that for any system of appreciable size, the probability of being in a state in which the energy is uniformly spread between the oscillators in overwhelmingly probable.  If we then begin with a system in a macrostate in which the energy is spread nonuniformly between the two solids and wait for the energy to rearrange itself, we are almost certain to find that the energy has exchanged to become more uniform among the solids (aside from small fluctuations).  The system then exhibits *irreversible* behaviour from purely probabilistic considerations. The energy exchange will stop near the most likely macrostate.  We call this "law of increase of multiplicity" the **second law of thermodynamics.**
###Large Systems
####Taxonomy of Numbers
A neat section classifying numbers:
**Small numbers** are typical numbers like 2, 10, or 42.
**Large numbers** come from exponentiating small numbers like $10^23$.  We can add a small number to a large number without altering it
$$10^{23} + 23 = 10^{23}.$$
**Very large numbers** come from exponentiating a small number by a large number, ie $10^{(10^{23})}$. We can multiply a very large number by a large number without altering it,
$$10^{(10^{23})}\cdot 10^{23} = 10^{10^{23} + 23} = 10^{(10^{23})}$$
####Stirlings Approximation

$$n! \approx n^n e^{-n} \sqrt{2\pi n}$$
$$\ln{n!} \approx n\ln{n} - n$$


####Multiplicity of Large Einstein Solid

For a large Einstein solid we can take N to be a large number such that,
$$\Omega(N,q)\approx {N+q \choose N} = \frac{(N+q)!}{N!q!}$$
We make this approximation as $\Omega$ will be a very large number and we can thus drop large number factors.

#####High-Temperature Limit
$q\gg N$ yields
$$\Omega(N,q) = \left(\frac{eq}{N}\right)^N$$

#####Low-Temperature Limit
We just switch the role of $N$ and $q$. For $q \ll N$
$$\Omega(N,q) = \left(\frac{eN}{q}\right)^q$$

####Sharpness of Multiplicity Function

If we take the multplicity of a two solid system in the high-temperature limit and expand around the peak, we find
$$\Omega = \Omega_{max}\cdot e^{-N(2x/q)^2}$$
so the multiplicity function has a peak of width,
$\sigma = \frac{q}{2\sqrt{N}}$
While this number is large, it is only a miniscule fraction of the total number of accessible microstates.  For reasonable numbers, the width is one ten-billionth of the full width of the space.  Near the edge, the multiplicity differs from the maximum by a factor of $e^{-100} \approx 10^{-44}$.

###Ideal Gas

####Multiplicity of an Ideal Gas

We begin with an ad hoc explanation (which will be even more ad hoc here than in Schroeder) which will follow into something more rigorous.

The multiplicity of a one particle gas will be,
$$\Omega_1 = \frac{VV_p}{h^3}$$
Where $V_p$ is really the surface area of a $d$ dimensional hypersphere with radius $r = \sqrt{2mU}$ that feeds from energy conservation. Interestingly, this creates an ambiguity in the definition of the entropy as to get the correct units, we must multiply by some energy scale $\Delta E$ that is the thickness of the hypershell.  As this is a multiplicative factor, it leads to an additive difference in the definition of the entropy.

For a gas of N particles, we have $\Omega_N = \frac{1}{N!}\frac{(VV_p)^N}{h^{3N}}$ where the $N!$ compensates for the overcounting of microstates related by exchanging identical particles. Inserting (using the trick involving a gaussian to get the d-dimensional solid angle. See the beginning of chapter 4 in Arovas' notes) and dropping some large factors yields,
$$\Omega_N \approx \frac{1}{N!}\frac{V^N}{h^{3N}}\frac{\pi^{3N/2}}{(3N/2)!}(\sqrt{2mU})^{3N} = f(N)V^NU^{3N/2}$$
We note now that the exponent of $U$ is half the total number of degrees of freedom, as it was for the Einstein solid.  These are special cases of a more general theorem that states, for a system with only quadratic degrees of freedom, the multiplicity is proportional to $U^{Nf/2}$.

####Interacting Ideal Gases

For interacting gases the multiplicity is
$$\Omega = [f(N)]^2(V_A V_B)^N(U_A U_B)^{3N/2}$$
The multiplicity function then has a peak of width $\frac{U_{total}}{\sqrt{3N/2}}$ and if $N$ is large, only a tiny fraction of the macrostates have a chance of occurring.

If we also allow the gases to exchange volume, the width of the peak as a function of $V_A$ is $\frac{V_{total}}{\sqrt{N}}$. For reference, if we sized the entire graph to fit on a page, the width of the peak would be less than an atom.

###Entropy

This suggests that any large system will be found in the macrostate with the highest multiplicity.  We form the entropy,
$$S = k_B \ln{\Omega}$$
and state this tendency as a law:

Any large system in equilibrium will be found in the macrostate with the greatest multplicity/entropy (aside from fluctuations that are normally too small to measure.)

or simply,

Entropy/Multiplicity tends to increase.

A nice consequence of our definition of entropy is that it makes it additive.  If we specify the macrostates of each system, then $S = S_A + S_B$. If these two systems can interact, we should instead compute the entropy over long time scales by finding the multiplicity of the whole system. As entropy is a function of the accessible microstates, it depends on the time scale, but in practice it rarely matters and if we just assume that the composite system is in its most likely macrostate, we get essentially the same entropy as if we summed over all of them.

####Entropy of Ideal Gas
From the multiplicity, employing Stirling's Approximation and throwing away some large factors we get the Sakur-Tetrode Equation:
$$S = Nk\left[\ln{\left(\frac{V}{N}\left(\frac{4\pi m U}{3Nh^2}\right)^{3/2} \right)}+\frac{5}{2}\right].$$
For a mole of helium at STP this gives

```python
h = 6.626e-34
def Sackur_Tetrode(N, V, m, U):
    """
    Sackur_Tetrode(N, V, m, U) returns the entropy of an ideal gas
    """
    return N*k_B*(np.log(V/N*(4*np.pi*m*U/(3*N*h**2))**(3./2))+5./2)
print("A mol of helium at STP has the entropy %0.1f J/K" % Sackur_Tetrode(N_A, R*T_room/atm, 0.004/N_A, 3./2*R*T_room))
```

##Interactions and Implications

###Temperature
Suppose we have two systems $A$ and $B$ that are in thermal contact so that they can exchange energy while the total energy of the system is constant. The system will exchange energy to bring the total entropy to a maximum.  As the entropy is additive,
$$\frac{\partial S}{\partial U_A} = 0 = \frac{\partial S_A}{\partial U_A} + \frac{\partial S_B}{\partial U_A}.$$
As the conservation of energy gives $dU_A = -dU_B$, we can arrange this to give that at equilibrium we will have,
$$\frac{\partial S_A}{\partial U_A} = \frac{\partial S_B}{\partial U_B}.$$
Examining the units here, we might draw the correspondence,
$$\left(\frac{\partial S}{\partial U}\right)_{N,V} = \frac{1}{T}$$
and this makes sense.  We should have that the system whose slope with respect to energy is larger will accept energy from the other as this will increase the *total* entropy of the system, and so the partial derivative with respect to energy should correspond to *inverse* temperature.

Rearranging we have $T = \left(\frac{\partial S}{\partial U}\right)_{N,V}$ which is not very useful analytically but may be useful numerically.  It is also worth noting that the temperature may not be precisely defined for small systems where the entropy is no longer smooth.

####High Temperature Einstein Solid
The entropy in the high temperature limit is $S = Nk\ln{\frac{eU}{\epsilon N}}$ This yields $\frac{1}{T} = \frac{Nk}{U}$ so $U=NkT$ agreeing with equipartition.

####Ideal Gas
From the Sackur-Tetrode Equation, $S = Nk\ln{U^{3/2}}+ ...$ so $\frac{1}{T} = \frac{3}{2}\frac{Nk}{U}$. This gives $U = \frac{3}{2}NkT$ again agreeing with equipartition.

####Low Temperature Einstein Solid
$S = \frac{Uk}{\epsilon}\ln{\frac{eN\epsilon}{U}}$ which gives $U = N\epsilon e^{-(\epsilon/kT)} = q\epsilon$ so $q = Ne^{-\epsilon/kT}$.  This is clearly nonsense for large $T$ as $q\to N$.  For small T, we get an exponential decrease in the number of quanta in the system.  This does not match experiment and will be fixed in Debye theory.

###Entropy and Heat
####Predicting Heat Capacities
To get experimentally verifiable results we take,
$$C_V = \left(\frac{\partial U}{\partial T}\right)_{N,V}$$
We get,
#####High Temperature Einstein Solid
$$C_V = Nk$$
#####Ideal Gas
$$C_V = \frac{3}{2}Nk$$

To get this we have followed the procedure:

1. Use quantum mechanics to find $\Omega(U,V,N,...)$
2. $S = k\ln{\Omega}$
3. $\frac{\partial S}{\partial U} = \frac{1}{T}$ gives $T(U,...)$
4. We invert this to get $U(T,...)$
5. $C_V = \frac{\partial U}{\partial T}$

####Measuring Entropy
We can measure the entropy of a substance by following this procedure in reverse:
We can calculate the change in entropy by measuring the heat transfer,
$$dS = \frac{dQ}{T} = \frac{C_V dT}{T}(constant\:volume,\:no\:work)$$
If $C_V$ is constant over the temperature range of interest, we have
$$\Delta S = C_V \ln{\frac{T_f}{T_i}}$$
If we know $C_V$ all the way to $T=0$,
$$S_f - S(0) = \int_0^{T_f}\frac{C_VdT}{T}$$
but what about $S(0)$? When the temperature goes to zero, we should have a single ordered ground state (assuming no degeneracy) so we should have $S\to0$ as $T\to0$ which is commonly called the **third law of thermodynamics**.  However in reality, things are not so convenient.

As a material approaches $T=0$ it typically becomes a solid (with the exception of helium which remains a liquid) and thus the atoms cannot rearrange easily after that.  Over time scales we care about, we thus have some frozen in *residual entropy*. Similarly, in solids we can change the orientation of molecules with very little change in the energy.  The rearrangement of these orientations is thus very slow and the entropy as $T\to0$ is $S_{residual} = k\ln{\left(\sharp\:of\:rearrangements\right)}$.  This may also come from mixing of nuclear isotopes throughout a material and the multiplicity of alignments of the spins which don't relax until T is very small.  Tabulated entropies usually do not include these.

#####Carbon Monoxide
For example, in solid carbon monoxide the energy difference between each molecule sitting as CO or OC is very small. This means that there are $2^N$ rearrangments of these molecules and
$$S_{resid} = Nk\ln{2}$$

We must note the divergence in $\int_0^{T_f}\frac{C_V}{T}dT$.  This means either $S_f\to\infty$ or $S(0)\to-\infty$ but both are
forbidden by $S=k\ln{\Omega}$ so we **must** have $C_V\to0$ as $T\to 0$ which is another version of the third law. This also means that the heat capacities we have calculated from equipartition cannot be correct for low temperatures.

####Macroscopic Entropy
$dS = \frac{dQ}{T}$ tells us how to calculate entropy but not what it is.  We note that entropy accompanies heat but how much depends on the temperature of the substance. At higher temperatures, the same heat causes a smaller change in the entropy.

Supposing that two solids, one at $300K$ and the other at $500K$ exchange $1500J$ of heat.  The changes in entropy are,
$$\Delta S_A = \frac{-1500J}{500K} = -3J/K$$
$$\Delta S_B = \frac{1500J}{300K} = 5J/K$$

###Paramagnetism
We consider a simple system to exercise the tools we have developed.  The two state paramagnet can be considered as N spin-1/2 particles in a constant magnetic field. The energy of a single dipole is proportional to the strength of the magnetic field and is at a minimum $-\mu B$ if the diplole points along the field and $\mu B$ if the dipole points opposite two the field.  As our particles are spin-1/2, these are the only states a dipole can occupy. The total energy of $N$ dipoles is thus (if the field is pointing up),
$$U = \mu B(N_\downarrow - N_\uparrow) = \mu B (N - 2N_\uparrow)$$
the magnetization is $M = \mu(N_\uparrow - N_\downarrow) = -\frac{U}{B}.$
The multiplicity thus takes the simple form,
$$\Omega = {N \choose N_\uparrow}$$

```python
def paramagnet_multiplicity(N, N_up):
    """
    paramagnet_multiplicity(N, N_up) - returns the multiplicity of a 2 state paramagnet with N_up of N dipoles in the
    up state
    """
    return scipy.special.binom(N, N_up)

N = 100
fig4, ax4 = subplots()
ax4.set_xlabel('Energy ($U/\mu$B)')
ax4.set_ylabel('Entropy (S/k)')
ax4.set_title('Entropy of a Two-State Paramagnet')
ax4.plot((N - 2*arange(N+1)), np.log(paramagnet_multiplicity(N*np.ones(N+1), arange(N+1))))
```

From the definition of temperature as inverse slope, we can see that the temperature goes to zero on the left, progresses to infinity as the energy goes to 0, goes 'through infinity' to negative infinity as the energy passes through zero and back to zero from the negative side as the energy becomes the highest possible value.

To understand this, it is more instructive to think in terms of $\frac{\partial S}{\partial U} = \beta$ than in terms of temperature.  When the energy is negative, the entropy has a positive slope and an increasing temperature and thus behaves normally.  When the energy goes to zero, the system will give up energy to any other system with positive slope.  After passing through zero, the system will still give up energy to any system with a slope that is higher than it's own, whether positive, negative, or zero.

We can thus think of negative temperatures as 'higher' than positive temperatures and when comparing two negative temperatures, the one closer to zero is higher and will tend to give up energy. Kindof silly, but that's the price of having higher temperatures give up energy I guess.

####Analytic Solutions

First, to get a workable form of the entropy, we invoke Stirlings approximation,

$$S/k = \ln{N!} - \ln{N_\uparrow !} -\ln{(N-N_\uparrow)!}$$
$$S/k = N\ln{N} - N_\uparrow\ln{N_\uparrow} - (N-N_\uparrow)\ln{N-N_\uparrow}$$

As $dU = -2\mu B dN_\uparrow$ we can re-express the derivative
$$\frac{1}{T} = -\frac{1}{2\mu B}\frac{\partial S}{\partial N_\uparrow}$$
$$\frac{1}{T} = \frac{k}{2\mu B}\ln{\left(\frac{N-U/\mu B}{N+U/\mu B}\right)}$$

Solving for U gives,
$$U = -N\mu B \tanh{\left(\frac{\mu B}{kT}\right)}$$
and the magnetization is
$$M = N\mu \tanh{\left( \frac{\mu B}{kT}\right)}.$$
Differentiating with respect to temperature gives the heat capacity:
$$C_B = \left(\frac{\partial U}{\partial T}\right)_{N,B} = Nk\cdot\frac{(\mu B/kT)^2}{\cosh^2{(\mu B/kT)}}$$

```python
bohr_mag = 9.27e-24

def paramagnet_magnetization(N, B, T):
    return N * bohr_mag * scipy.tanh(bohr_mag * B / (k_B * T))

def paramagnet_heatcapacity(N, B, T):
    x = bohr_mag * B / (k_B * T)
    return N * k_B * x**2 / scipy.cosh(x)**2

T = scipy.linspace(0, 10, 500)

fig5, axes5 = subplots(2,1, figsize=(8, 9))
axes5[0].plot(T, paramagnet_magnetization(N_A, 1, T))
axes5[0].set_title(r"Magnetization in 1 T with $N_A$ spins")
axes5[0].set_xlabel("Temperature (K)")
axes5[1].plot(T, paramagnet_heatcapacity(N_A, 1, T))

axes5[1].set_title(r"Heat Capacity")
axes5[1].set_xlabel("Temperature (K)")

fig5.tight_layout()
```

### Mechanical Equilibrium and Pressure

We can generalize the previous argument to the other partials of the entropy.  Consider the derivative with respect to volume,
$$\frac{\partial S_{total}}{\partial V_A} = 0 = \frac{\partial S_A}{\partial V_A} + \frac{\partial S_B}{\partial V_A} = 
\frac{\partial S_A}{\partial V_A} - \frac{\partial S_B}{\partial V_B}$$
yielding, at equilibrium
$$\left(\frac{\partial S_A}{\partial V_A}\right)_{U_A,N_A} = \left(\frac{\partial S_B}{\partial V_B}\right)_{U_B, N_B}.$$
Let's consider this expression carefully.  The partials here are taken with the volume and particle number fixed. Imagining a gas in a box separated by a moving wall. When the wall moves, doing work on each gas, some energy must be exchanged to keep the energies constant.  As the system is at equilibrium, we may also assume that the temperatures are equal.  We know that the system with greater pressure will tend to increase it's volume. Similarly, if $\frac{\partial S_A}{\partial V_A}$ is larger we would expect system $A$ to increase it's volume.  We thus suspect that
$\left(\frac{\partial S_A}{\partial V_A}\right)_{U_A,N_A} \propto P_A$ but the units don't quite balance.  Introducing a convenient factor of $T$ we identify
$$\left(\frac{\partial S_A}{\partial V_A}\right)_{U_A,N_A} = \frac{P_A}{T_A}.$$
Recalling that the entropy of an ideal gas only depends on $V$ in the logarithm ($S = Nk\ln{V} + f(N,U)$), $\frac{\partial S}{\partial V} = \frac{Nk}{V} = \frac{P}{T}$.  Excellent!  We could just as well consider this a derivation of the ideal gas law.

##### The Thermodynamic Identity

The relationships just derived may be summarized conveniently in the following form.  Take the differential of the entropy:
$$dS = \frac{\partial S}{\partial U} dU + \frac{\partial S}{\partial V} dV = \frac{dU}{T} + {P}{T} dV$$
which can be rearranged to form
$$dU = T dS - P dV$$
also known as the **thermodynamic identity**.  This equation is true for any infinitesimal change of a system and we can recover our other identities from it as well as $P = -\left(\frac{dU}{dV}\right)_S$.

##Boltzmann Statistics

###The Boltzmann Factor

By the fundamental assumption of statistical mechanics, every microstate of a system is equally probable.  If we split this system into a particular particle and a reservoir and specify the state of our particle, the probability of observing this state must be proportional to the multiplicity of the state of the reservoir. This means that the ratio of the probabilities of two states is equal to the ratio of the multiplicities:
$$\frac{P(s_1)}{P(s_2)} = \frac{\Omega_R(s_1)}{\Omega_R(s_2)}$$
We can rewrite the multiplicites in terms of the entropies:
$$\frac{\Omega_R(s_1)}{\Omega_R(s_2)} = \frac{e^{S_R(s_1)/k}}{e^{S_R(s_2)/k}} = e^{(S_R(s_1) - S_R(s_2))/k}$$
If the size of the reservoir is large compared to the particle (as it must be to be a reservoir), we can invoke the thermodynamic identity to find the change in the entropy.  If the system is only exchanging energy,
$$dS_R = \frac{dU_R}{T} = \frac{1}{T}(U_R(s_1) - U_R(s_2)) = -\frac{1}{T}(E(s_1) - E(s_2))$$
and so
$$\frac{P(s_1)}{P(s_2)} = \frac{e^{-E(s_1)/kT}}{e^{-E(s_2)/kT}}.$$
This tells us that the probability of finding the particle in a particular microstate of energy $E$ is,
$$P(s) = \frac{1}{Z}e^{-E(s)/kT}$$
with the partition function,
$$Z = \Sigma_s e^{-E(s)/kT}.$$

Note that we can always take the ground state to be $E_G=0$ as a shift in the energy levels is multiplicative in the Boltzmann factors and so cancels in the partition function. If we shift our energies in this way, at low temperatures $Z\approx 1$. The probability of observing a state is thus exponentially suppressed in $(E-E_G)/kT$.  At higher temperatures, the suppression is more severe.  To compute it, we can explicitly perform the sum, replace it with an integral approximation, or truncate it once the terms become small.

##Quantum Statistics

###The Gibbs Factor

If we repeat the argument used to derive the Boltzmann factors but now allow particle number to vary, we get,
$$dS_R = \frac{1}{T}(dU_R - \mu dN_R)$$
(assuming that the volume is fixed or that the energy of the volume change is so small that it can be safely neglected). This yields a new distribution referred to as the grand canonical ensemble,
$$P(s) = \frac{1}{\mathscr{Z}}e^{-[E(s) - \mu N(s)]/kT}.$$
The function $\mathscr{Z}(T) = \sum_s e^{-[E(s) - \mu N(s)]/kT}$ is the grand partition function or Gibbs sum and runs over all possible values of $E$ and $N$.

Note that if there is more than one species present we must include a $\mu_i N_i$ term for each.

###Bosons and Fermions

Gibbs factors are applied to dense systems in which two or more particles may want to occupy the same single particle state.  The trick is to consider our system to be the single particle state and the reservoir to be all other states. The probability of the state being occupied by $n$ particles is,
$$P(n) = \frac{1}{\mathscr{Z}}e^{-(n\epsilon - \mu n)/kT}= \frac{1}{\mathscr{Z}}e^{-n(\epsilon - \mu)/kT}$$
By considering the possible occupancies and calculating the average $\bar{n}$ we obtain:

#####Fermi-Dirac Distribution

$$\bar{n}_{FD} = \frac{1}{e^{(\epsilon-\mu)/kT}+1}$$

This goes to zero for $\epsilon \gg \mu$ and to 1 for $\epsilon \ll \mu$.  A state of energy $\mu$ always has a 50% chance of being occupied. We can remember the sign in the denominator because there is no divergence in $\bar{n}_{FD}$.

#####Bose-Einstein Distribution

$$\bar{n}_{BE} = \frac{1}{e^{(\epsilon-\mu)/kT}-1}$$

###Degenerate Fermi-Gas
For a gas of fermions at low enough temperature, the condition for Boltzmann statistics to apply,

Thus we are effectively at zero temperature for this gas and can look at slight variations from this to understand our normal physics.
